```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=["Group A", "Group B", "Group c", "Group D"]
yboys=[10,30,41,25]
xgirls=[50,45,60,50]

x_axis=np.arange(len(x))
plt.bar(x_axis -0.2, yboys, 0.4, label="boys")
plt.bar(x_axis +0.2, xgirls,0.4, label="girls")

plt.xticks(x_axis, x)
plt.xlabel("Groups")
plt.ylabel("Number of students")
plt.title("Number of students in each group")
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(10,5))
x=[12,3,4,5,2,3,4,5,6,3,4]
#print(len(x))
y=[2,3,5,6,4,7,5,6,3,1,2]
#print(len(y))

x1=[8,8,5,6,4,6,8,5,6,4,6]
y1=[9,6,4,5,8,5,6,8,9,6,4]
#print(len(y1))
#print(len(x1))

plt.scatter(x,y,label="high income low saving", color="r")
plt.scatter(x1,y1, label="low income high saving ", color="b")
plt.xlabel("Savings *100")
plt.ylabel("income*1000")
plt.title("scatter plot")
plt.legend(loc="best")
plt.show()
```

```python
#stackplot
fig=plt.figure(figsize=(10,5))
days=[1,2,3,4,5]

sleeping=[2,3,1,4,5]
eating=[8,9,6,3,4]
working=[6,4,7,8,9]
playing=[9,6,3,2,4]

plt.plot([],[],color='b',label="sleeping",linewidth=5)
plt.plot([],[],color='c',label="eating",linewidth=5)
plt.plot([],[],color='r',label="working",linewidth=5)
plt.plot([],[],color='g',label="playing",linewidth=5)

plt.stackplot(days,sleeping, eating, working, playing, colors=["b","c","r","g"])
plt.xlabel("x")
plt.ylabel("y")
plt.title("Stack plot")
plt.show()
```

```python
#pie chart
fig=plt.figure(figsize=(10,5))
#days=[1,2,3,4,5]

sleeping=[2,3,1,4,5]
eating=[8,9,6,3,4]
#working=[6,4,7,8,9]
#playing=[9,6,3,2,4]
#slices=[2,6,4,5,4]

activities=["sleeping","eating","working","playing","test"]
cols=["c","m","r","b",'k']
plt.pie(eating, 
        labels=activities,
        colors=cols, 
        startangle=90, 
        shadow=True, 
        explode=(0,0.1,0,0,0),
        autopct="%1.1f%%")
plt.title("Pie plot")
plt.show()
```

```python
#subplot
# 4 charts in  1 chart
fig=plt.figure(figsize=(10,5))
def f(t):
    return np.exp(-t) *np.cos(2*np.pi*t)
t1=np.arange(0,5.0,0.1)
t2=np.arange(0,5.0,0.2)
plt.subplot(221)
plt.bar([1,2,3,4],[3,2,4,1], color="r")
plt.legend(["first"])
plt.subplot(222)
plt.plot(t1,np.cos(2*np.pi*t1),color="g")
plt.legend(["second"], loc="best")
plt.subplot(223)
plt.plot(t1,f(t1),"r*", t2,f(t2))
plt.legend(["third"], loc="best")
plt.subplot(224)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.legend(["forth"], loc="best")
plt.show()
```

```python
#barh
fig=plt.figure(figsize=(10,5))
a=np.array([5,3,4,5,7,9])
b=np.array([9,2,4,7,5,6])
x=np.arange(6)
plt.barh(x, a,color="r")
plt.barh(x, -b,color="b")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.title("Bar h plot")
plt.show()
```

```python
#alpha
#grid
#align
a=np.array([3,6,9,4,2,5])
b=np.array([2,8,1,9,7,3])
x=np.arange(6)

plt.barh(x, a, align= "center", color="y")
plt.barh(x, -b, align= "center", color="r",zorder=3)

plt.grid(color="k")
plt.title("Back to back bar chart")
plt.xlabel("Values")
plt.ylabel("indexes")
plt.show()
```
